```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
time = np.arange(-1, 1,0.00001)
first_signal = np.cos(time * 10 * math.pi)
plt.plot(time,first_signal)
plt.xlabel("time")
plt.ylabel("Cos(10PI(time))")
plt.title("First Signal")
```

```python
def sampling_func(W,W1,main_signal,time):
    T = 2*math.pi / W
    T1 = 2*math.pi / W1
    sampeled_signal_time = time
    sampeled_signal = main_signal*0
    
    T_index = 0
    index_of_fisrt_zero = 0
    index_of_second_zero = 0
    first_zero_is_found = False
    
    for index in range(0,len(sampeled_signal_time)):
            if sampeled_signal_time[index] == 0 :   
                    index_of_fisrt_zero = index
                    break
    for index in range(index_of_fisrt_zero +1,len(sampeled_signal_time)):
            if int(10*(abs(time[index_of_fisrt_zero]) - abs(time[index])))  == int(10*T) :
                        print((time[index]))
                        index_of_second_zero = index
                        break
    
    FIRST_T_index = int((index_of_second_zero - index_of_fisrt_zero))
    print(FIRST_T_index)
    T_index = int(FIRST_T_index*T1/T)
    print(T_index)
    starting_point = 0
    for c in range(1,len(main_signal)) :
        if  starting_point + FIRST_T_index <= len(main_signal) :
            sampeled_signal[starting_point:starting_point +FIRST_T_index] += main_signal[starting_point:starting_point +FIRST_T_index]
            sampeled_signal[(starting_point - FIRST_T_index):starting_point] += main_signal[starting_point - FIRST_T_index:starting_point]
            starting_point += T_index
            
    plt.plot(time,sampeled_signal)
    plt.xlabel("time")
    plt.ylabel("Sampeled signal")
    plt.title("Plot Of Sampeled Signal")
```
