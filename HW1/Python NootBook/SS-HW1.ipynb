# (Part 1- a)

x1(t) = sin(t)

```python
Xa = np.arange(-math.pi,math.pi,0.01)
Ya = []
for number in Xa:
   Ya.append(math.sin(number))
plt.plot(Xa,Ya,'r')
plt.xlabel("t")
plt.ylabel("X1(t)")
plt.title("Sin(t)")
```

# (Part 1- b)

X2[n] = { -n-1 n > 0</br> 
pow(n,2) n <= 0 }

```python
Xb = np.arange(-5,6)
Yb = []
for number in Xb:
   if number < 0:
     Yb.append(-(number+1))
   else :
      Yb.append(math.pow(number,2))  
   
plt.bar(Xb,Yb,width = 0.3)
plt.xlabel("t")
plt.ylabel("X2(t)")
plt.title("Multi-Criteria-Function")
```

# (Part 1- c)

```python
Xc = np.arange(-5,11)
Yc = []

for number in Xc:
      if number >= -2 :
        if number == 0 :
            Yc.append(math.pow(2.71828,number/3) + 2)
        else :
            Yc.append(math.pow(2.71828,number/3))          
      else :
            Yc.append(0)
            
plt.bar(Xc,Yc, color = "black",width = 0.3)
plt.xlabel("t")
plt.ylabel("X3(t)")
plt.title("Exp and Step function")
```
