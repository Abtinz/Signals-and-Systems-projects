```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math
%reload_ext autoreload
%autoreload 2
```

```python
Xa = np.arange(-math.pi,math.pi,0.01)
Ya = []
for number in Xa:
   Ya.append(math.sin(number))
plt.subplot(1,3,1)
plt.plot(Xa,Ya,'r')
plt.xlabel("t")
plt.ylabel("X1(t)")
plt.title("Sin(t) : befor signal Transports")
#befor signal Transports
plt.subplot(1,3,3)
Xa = np.arange(-math.pi,math.pi,0.01)
Xa=Xa+3
Xa/=2
Ya = []
for number in Xa:
   Ya.append(math.sin(number))
plt.plot(Xa,Ya,'blue')
plt.xlabel("t")
plt.ylabel("X8(t)")
plt.title("Sin(t) : after signal Transports")
```

```python
Xa = np.arange(-math.pi,math.pi,0.01)
Ya = []
for number in Xa:
   Ya.append(math.sin(number))
plt.subplot(1,3,1)
plt.plot(Xa,Ya,'black')
plt.xlabel("t")
plt.ylabel("X1(t)")
plt.title("Sin(t) : befor signal Transports")
#befor signal Transports
plt.subplot(1,3,3)
Xa = np.arange(-math.pi,math.pi,0.01)
Xa=Xa-3
Xa/=-1
Ya = []
for number in Xa:
   Ya.append(math.sin(number))
plt.plot(Xa,Ya,'orange')
plt.xlabel("t")
plt.ylabel("X10(t)")
plt.title("Sin(t) : after signal Transports")
```

```python
Xc = np.arange(-5,11)
Yc = []

#n/3 --> Z(int) numbers
for number in Xc:
      if number >= -2 :
        if number == 0 :
            Yc.append(math.pow(2.71828,number*3) + 2)
        else :
            Yc.append(math.pow(2.71828,number*3))
                     
      else :
            Yc.append(0)
            
plt.stem(Xc,Yc)
plt.xlabel("t")
plt.ylabel("X3(t)")
plt.title("Exp Stepfunction\nbefor signal Transports")
```

```python
Xc1 = np.arange(2,17)/5
Xc1 *= -1
Xc = []
for number in Xc1:
    if number - int(number) == 0 :
        Xc.append(number)
Yc = []
for number in Xc:
      if number >= -2 :
        if number == 0 :
            Yc.append((math.pow(2.71828,number*3) + 2)*-2)
        else :
            Yc.append(math.pow(2.71828,number*3)*-2)        
      else :
            Yc.append(0)

plt.stem(Xc,Yc)
plt.xlabel("t")
plt.ylabel("X9(t)")
plt.title("Exp Stepfunction\nafter signal Transports")
```
