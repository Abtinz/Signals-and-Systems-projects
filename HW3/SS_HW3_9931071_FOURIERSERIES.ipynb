```python
import matplotlib.pyplot as plt
import numpy as np
```

# integral function

```python
def integ(signal):
    return np.sum(signal) * 0.001
```

# a(k) calculator

ùëéùëò = 2/ùëá0 * ‚à´ ùë•(ùë°)ùëêùëúùë†(ùëòùúî0ùë°) ùëët

```python
def ak_calculator(signal, k, T0,w0):
    integrable_signal = np.cos(k*w0*t) * signal
    return (2/T0) * integ(integrable_signal)
```

# b(k) calculator

ùëèùëò = 2/ùëá0 ‚à´ ùë•(ùë°) * ùë†ùëñùëõ(ùëòùúî0ùë°) ùëëùë° 

```python
def bk_calculator(signal, k, T0,w0):
    integrable_signal = signal * np.sin(k*w0*t)
    return (2/T0) * integ(integrable_signal)
```

# Fourier Series

ùë•(ùë°) = ùëé0/2 + ‚àë(ùëò=1..‚àû) ùëéùëò ùëêùëúùë†(ùëòùúî0ùë°) + ‚àë(ùëò=1..‚àû) ùëèùëò ùë†ùëñùëõ(ùëòùúî0ùë°)  

```python
def fourier_series(signal,time, c, T0):
    w0 = 2*np.pi/T0
    y_signal = np.zeros((len(time)))
    y_signal += ak_calculator(signal, 0, T0,w0)/2
    for k in range(1, c + 1):
        y_signal += ak_calculator(signal, k, T0 ,w0)* np.cos(k*w0*time) + bk_calculator(signal, k, T0,w0) * np.sin(k*w0*time)
    return y_signal
```

```python
time = np.arange(-6, 6,0.01)
signal =np.arange(-6, 6,0.01)*0
for index in range(0,len(time)) :
    if time[index] <= -3 or ( time[index] >=0 and time[index]<=3):
        signal[index] = 1

plt.plot(time,signal)
plt.title("Crucial Signal")
plt.xlabel("time")
plt.ylabel("signal")
```

# Plot Figure

In this function, with 5 different c's that have coefficients k in the Fourier series, we draw each series of 5 graphs to check the convergence of the coefficients.

```python
def plot_figure(signal,time,c_range):
    plotFigure = plt.figure(figsize=(5,5))
    axes = plotFigure.add_axes([1,1,1,1])
    axes.set_title("fourir series function")
    T0 = 6
    for c in range(c_range,c_range+5):
         Y = fourier_series(signal,time, c, T0)
         lable_c = ">>" + str(c) + "<<"
         if c % 5 == 0 :
              axes.plot(time,Y,"black",label = lable_c)
         elif c % 5 == 1:
              axes.plot(time,Y,"r",label = lable_c)
         elif c % 5 == 2:
              axes.plot(time,Y,"blue",label = lable_c)
         elif c % 5 == 3:
              axes.plot(time,Y,"green",label = lable_c)
         else:
              axes.plot(time,Y,"yellow",label = lable_c)

    axes.legend(loc=0)
```

```python
plot_figure(signal,time,0)
```

```python
plot_figure(signal,time,5)
```
