```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def integ(signal):
    return np.sum(signal) * 0.01
```

```python
def fourier_transform(signal,time):
    w0 = 2*np.pi/30
    multipled_signal_with_exp = []
    time_index = 0
    for number in signal:
        multipled_signal_with_exp.append(number*math.exp(-w0*time[time_index]))
        time_index +=1

    return multipled_signal_with_exp
```

```python
time = np.arange(-15, 15,0.01)
first_signal =np.arange(-15, 15,0.01)*0
for index in range(0,len(time)) :
    if time[index] <= 1 and time[index] >=-1 :
        first_signal[index] = 1

plt.plot(time,first_signal)
plt.title("first Signal")
plt.xlabel("time")
plt.ylabel("signal")
```

```python
second_signal =np.arange(-15, 15,0.01)*0
for index in range(0,len(time)) :
        second_signal[index] = math.sin(time[index]) / (math.pi * time[index])

plt.plot(time,second_signal)
plt.title("second Signal")
plt.xlabel("time")
plt.ylabel("signal")
```

```python
third_signal =np.arange(-15, 15,0.01)*0
for index in range(0,len(time)) :
        third_signal[index] =(math.cos(3*time[index])  * math.sin(3*time[index]) ) / (math.pi * time[index])

plt.plot(time,third_signal)
plt.title("third Signal")
plt.xlabel("time")
plt.ylabel("signal")
```

```python
fourth_signal =np.arange(-15, 15,0.01)*0
for index in range(0,len(time)) :
        fourth_signal[index] = math.pow(second_signal[index],2)

plt.plot(time,fourth_signal)
plt.title("fourth Signal")
plt.xlabel("time")
plt.ylabel("signal")
```

```python
plt.plot(time,fourier_transform(first_signal,time) , 'r')
plt.title("fourier transform of first Signal")
plt.xlabel("frecanse")
plt.ylabel("signal")
```

```python
plt.plot(time,fourier_transform(second_signal,time), 'r')
plt.title("fourier transform of second Signal")
plt.xlabel("frecanse")
plt.ylabel("signal")
```
