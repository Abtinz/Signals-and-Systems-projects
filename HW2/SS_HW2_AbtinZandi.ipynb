```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math
```

# Convolution Function

this function will convolve x(t) and h(t) then it will return

```python
def convolution_function (x_first_signal , t_first_signal , x_second_signal , t_second_signal,signal_type):
    biggest_sift =  2*(t_first_signal[len(t_first_signal)-1] + t_second_signal[len(t_second_signal)-1])
    t_second_signal = t_second_signal * -1
    n_shift = []
    #posetive
    current_number = 0
    while True :
        signal_passed = False
        for index in range(0,len(t_second_signal)):
            if t_second_signal[index] + current_number in t_first_signal:
                    signal_passed = True
                    n_shift.append(current_number)
                    break
        if signal_passed:
            current_number -=-1
            
        else :
            break
    #negative
    current_number = -1
    while True :
        signal_passed = False
        for index in range(0,len(t_second_signal)):
            if t_second_signal[index] + current_number in t_first_signal:
                    signal_passed = True
                    n_shift.append(current_number)
                    break
        if signal_passed:
            current_number -=1
            
        else :
            break

    convolution_resault = []
    for number in n_shift:
        current_y = 0
        for index in range(0,len(t_second_signal)):
            for second_index in range(0,len(t_first_signal)):
                if t_second_signal[index] + number == t_first_signal[second_index]:
                    current_y += t_second_signal[index] * t_first_signal[second_index]
                    break
        
        convolution_resault.append(current_y)
    
    plot_legend(n_shift,convolution_resault,signal_type)
       
    
        
```

```python
def  plot_legend(n_shift,convolution_resault,signal_type): 
    if signal_type:
        plt.plot(n_shift,convolution_resault,'black')
        plt.xlabel("t")
        plt.ylabel("Y(t)")
        plt.title("convolution signal") 
    
    else:
        plt.stem(n_shift,convolution_resault,'blue')
        plt.xlabel("n")
        plt.ylabel("Y[n]")
        plt.title("convolution signal") 
```

```python
T_X1 = np.arange(0,10.01 , 0.01)
X1 = np.arange(0,10.01 , 0.01,dtype=int)*0 - 1

plt.plot(T_X1,X1,'r')
plt.xlabel("t")
plt.ylabel("X1(t)")
plt.title("First Signal --> u(t - 10) - u(t)")
```

```python
T_X2 = np.arange(0,10.01 , 0.01)
X2 = []
for number in T_X2:
    if number < 5 :
        X2.append(number)
    else :
        X2.append(5-number)
plt.plot(T_X2,X2,'blue')
plt.xlabel("t")
plt.ylabel("X2(t)")
plt.title("Second Signal")
```

```python
plotFigure = plt.figure(figsize=(5,5))
axes = plotFigure.add_axes([1,1,1,1])
axes.set_title("X1 and X2 function")
axes.plot(T_X1,X1,"black",label = "X1")
axes.plot(T_X2,X2,"r",label = "X2")
axes.legend(loc=0)
```

```python
convolution_function(X1 ,T_X1 , X2 , T_X2,True )
```

```python
t_h = np.arange(0,5)

h_signal = np.arange(0,5 , dtype = float)*0 - 1
for index in range(0,5):
    h_signal[index] *= math.pow(0.9,index)
plt.stem(t_h,h_signal,'blue')
plt.xlabel("n")
plt.ylabel("h[n]")
plt.title("h Signal")
```
